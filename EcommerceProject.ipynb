```python
clv=df.groupby('customer_unique_id')['total_price'].sum().reset_index()
clv.columns = ['customer_id', 'lifetime_value']
```

```python
frequency=df.groupby('customer_unique_id')['order_id'].nunique().reset_index()
frequency.columns = ['customer_id', 'frequency']
frequency
```

```python
clv
```

```python
customer_data = clv.merge(frequency, on='customer_id')
customer_data
```

```python
def segment(x):
    if x > 1000:
        return 'High Value'
    elif x >= 300:
        return 'Medium Value'
    else:
        return 'Low Value'
```

```python
monthly_sales = df.groupby(['year','month'])['total_price'].sum().reset_index()
monthly_sales
```

```python
category_sales = df.groupby('product_category_name')['total_price'].sum().reset_index()
category_sales
```

```python
location_sales = df.groupby('customer_city')['total_price'].sum().reset_index()
location_sales
```
